```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
```

```python
iris_filepath = ('/home/student/Downloads/Iris.csv')
iris_data = pd.read_csv(iris_filepath)
iris_data.head()
```

```python
iris_data.dtypes
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

sns.histplot(data = iris_data, x = 'SepalLengthCm', bins = 20, ax = axes[0, 0])
axes[0, 0].set_title('Histogram for Sepal Length')

sns.histplot(data = iris_data, x = 'SepalWidthCm', bins = 20, ax = axes[0, 1])
axes[0, 1].set_title('Histogram for Sepal Width')

sns.histplot(data = iris_data, x = 'PetalLengthCm', bins = 20, ax = axes[1, 0])
axes[1, 0].set_title('Histogram for Petal Length')

sns.histplot(data = iris_data, x = 'PetalWidthCm', bins = 20, ax = axes[1, 1])
axes[1, 1].set_title('Histogram for Petal Width')


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

sns.boxplot(data = iris_data, x = 'Species', y = 'SepalLengthCm', ax = axes[0, 0])
axes[0, 0].set_title('Iris Sepal Length Boxplot')

sns.boxplot(data = iris_data, x = 'Species', y = 'SepalWidthCm', ax = axes[0, 1])
axes[0, 1].set_title('Iris Sepal Width Boxplot')

sns.boxplot(data = iris_data, x = 'Species', y = 'PetalLengthCm', ax = axes[1, 0])
axes[1, 0].set_title('Iris Petal Length Boxplot')

sns.boxplot(data = iris_data, x = 'Species', y = 'PetalWidthCm', ax = axes[1, 1])
axes[1, 1].set_title('Iris Petal Width Boxplot')

plt.tight_layout()

plt.show()
```
